## ✅ .corr()

- ```DataFrame.corr(method='pearson', min_periods=1, numeric_only=False)```
- DataFrame 열 간의 쌍별 ⭐상관관계⭐를 계산하는 함수

---

- 매개변수
    - `method: {'pearson', 'kendall', 'spearman'}` 또는 호출 가능한(callable) 함수, 기본값: `'pearson'`
    - = 상관관계 계산에 사용할 방법 지정
        - `'pearson`' : 표준 상관 계수
        - `'kendall'` : Kendall Tau 상관 계수
        - `'spearman'` : Spearman 순위 상관 계수
        - `callable` : 두 개의 1차원 ndarray를 입력으로 받아 float를 반환하는 함수.
        - ⏬
    
    - `min_periods`: `int`, 선택 사항, 기본값: `1`
    - = 유효한 설명을 얻기 위해 각 열 쌍에 필요한 최소 관측치 수
        - ⏬
    - `numeric_only`: `bool`, 기본값: `False`
    - = `float`, `int` 또는 `boolean` 데이터만 포함할지 여부를 지정

- 반환값:
    - `DataFrame`
    - = 상관 행렬 반환

````python
df.corr(numeric_only=True)  # 숫자형, 불리언 포함

# 원핫 인코딩된 범주형 변수가 있다면
# 인코딩된 열의 값과 다른 변수와 상관관계 확인 가능
````
