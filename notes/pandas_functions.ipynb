## ‚úÖ .corr()

- ```DataFrame.corr(method='pearson', min_periods=1, numeric_only=False)```
- DataFrame Ïó¥ Í∞ÑÏùò ÏåçÎ≥Ñ ‚≠êÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ‚≠êÎ•º Í≥ÑÏÇ∞ÌïòÎäî Ìï®Ïàò

    - = ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ Í≥ÑÏÇ∞Ïóê ÏÇ¨Ïö©Ìï† Î∞©Î≤ï ÏßÄÏ†ï
        - `'pearson`' : ÌëúÏ§Ä ÏÉÅÍ¥Ä Í≥ÑÏàò
        - `'kendall'` : Kendall Tau ÏÉÅÍ¥Ä Í≥ÑÏàò
        - `'spearman'` : Spearman ÏàúÏúÑ ÏÉÅÍ¥Ä Í≥ÑÏàò
        - `callable` : Îëê Í∞úÏùò 1Ï∞®Ïõê ndarrayÎ•º ÏûÖÎ†•ÏúºÎ°ú Î∞õÏïÑ floatÎ•º Î∞òÌôòÌïòÎäî Ìï®Ïàò.
        - ‚è¨
    
    - `min_periods`: `int`, ÏÑ†ÌÉù ÏÇ¨Ìï≠, Í∏∞Î≥∏Í∞í: `1`
    - = Ïú†Ìö®Ìïú ÏÑ§Î™ÖÏùÑ ÏñªÍ∏∞ ÏúÑÌï¥ Í∞Å Ïó¥ ÏåçÏóê ÌïÑÏöîÌïú ÏµúÏÜå Í¥ÄÏ∏°Ïπò Ïàò
        - ‚è¨
    - `numeric_only`: `bool`, Í∏∞Î≥∏Í∞í: `False`
    - = `float`, `int` ÎòêÎäî `boolean` Îç∞Ïù¥ÌÑ∞Îßå Ìè¨Ìï®Ìï†ÏßÄ Ïó¨Î∂ÄÎ•º ÏßÄÏ†ï
