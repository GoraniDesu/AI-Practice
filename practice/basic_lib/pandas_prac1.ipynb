## 1) 데이터 로딩

```python
import kagglehub
import pandas as pd
# Download latest version
path = kagglehub.dataset_download("melihkanbay/police")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path+'/police.csv')
```

```python
df.head()
```

## 2) 데이터 확인

```python
df.info()
```

```python
# 데이터가 없는 열 삭제하기
police = df.copy()
police = police.drop(labels="county_name", axis=1)

police.info()
```

## 3) 데이터 색인
- 2005년 1월 데이터만 보기
- 남성인 데이터만 보기
- 20세인 데이터만 보기

```python
print("==== 2005년 1월인 데이터 ====")
display(police[police['stop_date'].str.startswith("2005-01")])

print("\n\n==== 남성인 데이터 ====")
display(police[police['driver_gender'] == "M"])

print("\n\n==== 20세 데이터 ====")
display(police[police['driver_age'] == 20])
```

- 결과
  - 2005년 1월 데이터: 3개
  - 남성인 데이터: 62895개
  - 20세 데이터: 3794개

## 4) 데이터 연산
- 발생 이후 경과 일수
- 현재 - 발생일

```python
print("==== stop_date의 Dtype 확인하기 ====\n")
police['stop_date'] = pd.to_datetime(police['stop_date'], yearfirst=True, exact=True)
print(police.info())

print("\n\n==== 상위 3개 확인하기 ====")
display(police.head(3))
```

```python
police['days_since_event'] = pd.to_datetime("today")-police["stop_date"]
police.iloc[:, [0, -2, -1]]
```
